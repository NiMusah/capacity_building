The job details section with informtion about Kasha as a company

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

kasha_about='''Kasha is a fast-growing startup ecommerce company in East Africa focused on womens health 
    and personal care, selling products such as menstrual care, contraceptives, soaps, lotions, baby diapers, 
    and beauty products.
    Kashas ecommerce is highly accessible across the region and customers can order via the website, app, 
    USSD (basic mobile phone, no internet), call center or social media. On top of the consumer business, Kashas 
    business includes market insight, analytics and service offerings to global manufacturers of FMCG, Pharma and 
    Beauty products with the aim of helping them drive their revenue and market penetration.
    We are a purposedriven company, and our mission is to disrupt and democratize access to womens health 
    products, services and information for women of all socioeconomic levels in urban and rural areas of developing 
    countries.
    The company was founded in July 2016 in Rwanda and expanded into Kenya in the first half of 2018. Kasha 
    strives to one day serve millions of women around the world in emerging markets.'''

kasha_about_cloud=WordCloud().generate(kasha_about)
plt.imshow(kasha_about_cloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
