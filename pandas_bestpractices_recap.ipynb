```python
#Dummy encoding can be enforced using the panda's get_dummies function as well but with an addition parameter/arguement.
#Dummmy encoding has the characteristic of droping the first instance of the categorical column value
#In this sense it creates n-1 features for n-categories.
#As you will notice, from the results, the 'race_Asian' feature will be dropped compared to the one-hot encoding where it
#was inclided
#race_Asian acts as the base value and is encoded by the absence of all other categorical values
#In this sense the last feature 'race_White' will represent the base value as an intercept
```

```python
driver_race_dummy=pd.get_dummies(only_objects, columns=['driver_race'], drop_first=True, prefix='race')
print(driver_race_dummy.head(10))
```

```python
#To counter the challenges(instability e.g.) each of the encoding types brings it might be a good idea to limit your columns
#Encoding may result in a huge number of columns if there are huge number of categories in a categorical column
#To limit this, the value_count function can be relied on to only create columns for the most categorical values
#The value_count function counts the occurences of each categorical value
#As the results will show, race_White has the number of occurences
#Once we get the values, we can limit what values will be included in the encoding by first creating a mask of values
#These values are those that occur less than n-times
```

```python
#A mask in this sense will be a list of booleans outlining which values in a column should be affected
#We first specify the categorical column and the values that occur less than n-times using the index attribute
#We then wrap this inside the isin method
#After creating the mask we can use it to replace the categories that occur less than n-times with a value of your choice
```

```python
counts=ri1['driver_race'].value_counts()
print(counts)
```

```python
mask=ri1['driver_race'].isin(counts[counts<10000].index)
ri1['driver_race'][mask]='Other'
print(pd.value_counts(mask))
```
