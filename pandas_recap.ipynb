1. Reading Tabular Data Files into Pandas

```python
#The pandas read_table function can read a url directly
```

```python
import pandas as pd
chipotle_dataset=pd.read_table('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/chipotle.tsv')
chipotle_dataset.head()
```

```python
#The chipotle dataset we just used is soemhow clean and can be worked on with minimum data manipulation
#However, the dataset movie_users (demographic info) is not exactly clean and requires a little bit of manipulation
```

```python
movie_users=pd.read_table('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/u.user')
movie_users.head()
```

```python
#As seen from the above result, the dataset pulled lack a proper seperator/delimitor and an argument has to provided 
#The arguement is the pandas sep/delimiter arguement
#In this case, the delimiter/sep to be specified is the pipe('|') delimiter
#This will result in a dataframe with the proper seperators/delimiters 
```

```python
movie_users_url=('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/u.user')
movie_users=pd.read_table(movie_users_url, sep='|')
movie_users.head()
```

```python
#From the above result, we find that the dataset still needs some work
#The resulting dataset/frame displayus the first row as a column header which is not the case
#We therefore have to undertake data manipulation to correct this
#To do that, we have to specify the header arguement which in this case will be none
#It is None becasue we do not want the first row being interpreted as a column header
```

```python
movie_users=pd.read_table(movie_users_url, sep='|', header=None)
movie_users.head()
```

```python
#What happens now is that specifying the header arguemnt as None means pandas will use default integer values as seen above
#However, this is still not ideal and we have to define column names for our dataset/frame, however you want to call it
#We can do this by creating a list (movie_users_cols) of desired column header values which will then be passed
#Passing the list of column names to the dataframe can be achieved through the use of the names arguement
```

```python
movie_users_url=('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/u.user')
movie_users_cols=['user_id', 'age', 'gender', 'occupation', 'zip_code']
movie_users=pd.read_table(movie_users_url, sep='|', header=None, names=movie_users_cols)
movie_users.head()
```
